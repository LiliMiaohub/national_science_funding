```python
import pandas as pd
```

```python
funder_df = pd.read_csv("../../../data/Data/RawData/article_funder_country_no_multi.csv",
                        sep="\t",error_bad_lines=False, warn_bad_lines=False,encoding="utf-16")
funder_df.head()
```

```python
pubs_funder = funder_df.groupby('ID_Art')['Country_new'].apply(list).reset_index()
pubs_funder.head()
```

```python
author_df = pd.read_csv("..//..//..//data//Data//RawData//article_author_org_country_discint.csv", 
                                sep=";",error_bad_lines=False, warn_bad_lines=False)
author_df.head()
```

```python
pubs_author = author_df.groupby('ID_Art')['author_country'].apply(list).reset_index()
pubs_author.head()
```

```python
pubs_author_funder = pubs_author.merge(pubs_funder, on="ID_Art", how="inner")
pubs_author_funder.columns=['id','author_country','funder_country']
pubs_author_funder.head()
```

```python
pubs_author_funder.to_csv("..//..//..//data//Data//DerivedData//CleanedRawData//pubs_author_funder.csv", index=False)
```

```python
import pandas as pd
```

```python
nofund_df = pd.read_csv("../../../data/Data/RawData/non_fund_pub_author_country.csv",
                       sep="\t",error_bad_lines=False, warn_bad_lines=False,encoding="utf-16")
nofund_df.head()
```

```python
article_nofund=nofund_df.groupby(
    ['ID_Art','Pb_year','Code_Discipline'])['author_country'].apply(list).reset_index()
article_nofund.head()
```

```python
nofund_detail=pd.read_csv("../../../data/Data/RawData/non_funded_article_details.csv",
                         sep=";",error_bad_lines=False,warn_bad_lines=False,encoding="utf-16")
nofund_detail.head()
```

```python
nofund_filter=nofund_detail[['ID_Art','Cit_Rel_ALL','Cit_ALL']].drop_duplicates()
nofund_filter.head()
```

```python
nofund_filter.shape
```

```python
article_nofund.shape
```

```python
article_nofund = article_nofund.merge(nofund_filter,on="ID_Art",how="inner")
article_nofund.head()
```

```python
article_nofund.columns=['id','year','discipline','author_country','rel_citation','citation']
article_nofund.head()
```

```python
funder_df = pd.read_csv("../../../data/Data/RawData/article_funder_country_no_multi.csv",
                        sep="\t",error_bad_lines=False, warn_bad_lines=False,encoding="utf-16")
funder_df.head()
```

```python
pubs_funder = funder_df.groupby('ID_Art')['Country_new'].apply(list).reset_index()
pubs_funder.head()
```

```python
fund_author = pd.read_csv("..//..//..//data//Data//RawData//article_author_org_country_discint.csv", 
                                sep=";",error_bad_lines=False, warn_bad_lines=False)
fund_author.head()
```

```python
fund_author = fund_author.groupby('ID_Art')['author_country'].apply(list).reset_index()
fund_author.head()
```

```python
fund_author_funder = fund_author.merge(pubs_funder, on="ID_Art", how="inner")
fund_author_funder.head()
```

```python
fund_author_funder.columns=['id','author_country','funder_country']
fund_author_funder.head()
```

```python
meta_df = pd.read_csv("..//..//..//data//Data//RawData//ID_Art_funder_author_country_flux.csv",
            sep="\t",encoding="utf-16",error_bad_lines=False)
meta_df.head()
```

```python
meta_df=meta_df[['ID_Art','Code_Discipline','Pb_Year','Cit_Rel_ALL','Cit_ALL']]
meta_df.columns=['id','discipline','year','rel_citation','citation']
meta_df.head()
```

```python
meta_df = meta_df.drop_duplicates()
meta_df.head()
```

```python
meta_df.shape
```

```python
fund_author_funder.shape
```

```python
fund_author_funder = fund_author_funder.merge(meta_df, on="id",how="left")
fund_author_funder.head()
```

```python
article_nofund.head()
```

```python
pubs_all = pd.concat([fund_author_funder,article_nofund])
```

```python
pubs_all
```

```python
def decm(s):
    ss = s.split(",")
    if len(ss[0])==0:
        s=s.replace(",","0.")
    else:
        s=s.replace(",",".")
    return s 
```

```python
pubs_all['rel_citation']=pubs_all['rel_citation'].apply(lambda x:decm(x))
pubs_all.head()
```

```python
pubs_all.to_csv("..//..//..//data//Data//DerivedData//CleanedRawData//pubs_author_funder.csv",
               index=False)
```

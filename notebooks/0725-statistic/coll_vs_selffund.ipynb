#### this script measures the amount of collaboration for each country and the proportion of self-funding of each country

```python
import pandas as pd
```

```python
pubs_path="../../data/NationalFunding/Data/DerivedData/CleanedRawData/pub.pkl"
```

```python
df = pd.read_pickle(pubs_path)
```

```python
df.head()
```

```python
int_df = df[['author_distinct','IntCol']].explode('author_distinct')
int_df = int_df.groupby(by=['author_distinct','IntCol']).size().reset_index(name='count')
int_df = int_df.pivot(index='author_distinct',columns='IntCol',values='count')
```

```python
int_df.rename(columns={0:'not',1:'col'},inplace=True)
```

```python
int_df['total'] = int_df['not']+int_df['col']
int_df['col_p'] = int_df['col']/int_df['total']
int_df = int_df.sort_values(by='col_p',ascending=False)
```

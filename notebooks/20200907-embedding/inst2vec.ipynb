```python
import gensim
import pandas as pd
```

```python
nofund_df = pd.read_csv("../../../data/Data/RawData/non_fund_pub_author_country.csv",
                       sep="\t",error_bad_lines=False, warn_bad_lines=False,encoding="utf-16")
nofund_df.head()
```

```python
nofund_inst=nofund_df.groupby(
    'ID_Art')['Institution'].apply(lambda x:list(set(x))).reset_index()
nofund_inst.head()
```

```python
fund_df = pd.read_csv("../../../data/Data/RawData/article_funder_country_no_multi.csv",
                        sep="\t",error_bad_lines=False, warn_bad_lines=False,encoding="utf-16")
fund_df.head()
```

```python
fund_finst=fund_df.groupby(
    'ID_Art')['Organisation'].apply(lambda x:list(set(x))).reset_index()
fund_finst.columns=['id','fund_inst']
fund_finst
```

```python
fund_author = pd.read_csv("..//..//..//data//Data//RawData//article_author_org_country_discint.csv", 
                                sep=";",error_bad_lines=False, warn_bad_lines=False)
fund_author.head()
```

```python
fund_ainst=fund_author.groupby(
    'ID_Art')['Institution'].apply(lambda x:list(set(x))).reset_index()
fund_ainst.columns=['id','author_inst']
fund_ainst.head()
```

```python
fund_finst.shape
```

```python
fund_inst=fund_ainst.merge(fund_finst,on="id")
fund_inst
```

```python

nofund_inst.columns=['id','author_inst']
nofund_inst.head()
```

```python
pubs_inst = pd.concat([fund_inst,nofund_inst])
```

```python
pubs_inst.to_csv("pubs_inst.csv",index=False)
```

```python
pubs_inst
```

```python
tokens=pubs_inst[pubs_inst.fund_inst.notnull()].fund_inst.tolist()
```

```python
model = gensim.models.Word2Vec(
            tokens,
            size = 100,
            window = 10, # just use the entire sentence
            min_count = 5, # Remove tokens that don't appear enough
            workers = 4, # paralellize, use 4 workers
            iter = 1,
            sg = 1 # use the skip_gram model
) # end model
```

```python
model.most_similar("NIH")
```

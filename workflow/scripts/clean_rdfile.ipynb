this script cleans oecd rd data

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
rd_path="../../data/nf_folder/Data/AdditionalData/GDEXPRD_OECD.csv"
namemap_path="../../data/nf_folder/Data/AdditionalData/country_WoS_ECI_WB_flags.tsv"
```

```python
rd_cleaned_path='../../data/nf_folder/Data/DerivedData/CleanedRawData/GDEXPRD_OECD.csv'
```

```python
rd_df=pd.read_csv(rd_path)
namemap_df=pd.read_csv(namemap_path,sep="\t")
```

```python
rd_df=rd_df.merge(namemap_df[['Code','WoS']],left_on='LOCATION',right_on='Code',how='left')
```

print out countries that are included in r&d file but are not covered by wos

```python
removelist = rd_df[rd_df['WoS'].isna()].LOCATION.unique()
removelist
```

remove these two countries from r&d file

```python
rd_df.head()
```

```python
rd_df=rd_df[~rd_df.LOCATION.isin(removelist)]
rd_df.head()
```

```python
rd_df=rd_df.rename(columns={'INDICATOR':'indicator','SUBJECT':'subject',
                           'MEASURE':'measure','TIME':'time','Value':'value'})
rd_df.head()
```

```python
rd_df[['Code','WoS','time','indicator','subject','measure','value']].to_csv(rd_cleaned_path,index=False)
```
